In your new code cell below, write a function `bigger` with the following specifications.
* The arguments to `bigger` are two real numbers.
* `bigger` prints to screen the larger of the two numbers. If the two numbers are equal, then print that number.
* As output, return the difference between the larger number and the smaller number.

```python
def bigger(realNum1, realNum2):
    #Check first if the two numbers are equivalent.
    if (realNum1 == realNum2):
        #Print either of them since they're equal (I just picked the second num but it doesn't matter)
        print(realNum2)
        #Return at the end
        return 0
    #If they're not equal...
    else:
        #Find the smaller of the two numbers
        smallNum = min(realNum1, realNum2)
        #Find the larger of the two numbers
        bigNum = max(realNum1, realNum2)
        #Find the difference.
        diff = (bigNum - smallNum)
        #Print the bigger number.
        print(bigNum)
        #Return the difference.
        return diff
```

### (2a) 

Implement a class `Jedi`, with the following specifications.
* A new instantiation of `Jedi` takes in a single argument for the `name` of the Jedi.
* New Jedis have an attribute `skills`, which will be a list of the Jedis learned skills.
* New Jedis have a Boolean attribute `isGhost`
* New Jedis have a character string attribute `name`, which is assigned the value passed into the instantiation.
* New Jedis are not ghosts and have no skills.

Jedi have the following methods.
* `learn(char string)`: push a new skill to the end of the Jedi's list of skills
* `forget()`: sometimes Jedis forget things too! Pop off the last skill the Jedi learned. If the Jedi has no skills, return "Jedi has no skills!"
* `checkSkills()`: return an integer representing the number of unique skills the Jedi has
* `duel(other_jedi)`: the given Jedi duels another Jedi (`other_jedi`), who is input to the `duel` method. The winner of the duel is the Jedi who possesses more (unique) skills. The loser of the duel turns into a ghost. This method should return the name of the winner of the duel. If the duel is a tie, return "Tie". Additionally, your `duel` routine should:
  * Print "[name] wins the duel.", where `[name]` is replaced by the name of the winning Jedi.
  * If there is a tie, print "The duel results in a tie."

**SOURCES USED:**

1 - (Set Method) https://www.geeksforgeeks.org/python-set-method/

2 - (If/Else in Classes using Self) https://stackoverflow.com/questions/40810139/how-to-write-python-if-else-in-single-line

```python
class Jedi:
    #First Part
    #----------------------------------
    #Instantiation of Jedi. Takes single argument name. So pass in self, and name.
    def __init__(self, name):
        #Attribute of skills is a list -> Empty list.
        self.skills = []
        #A boolean for "is it a ghost?" -> Set to False. They're not Ghosts.
        self.isGhost = False
        #String attribute called "name" which is passed in above.
        self.name = name
        
    #Second Part
    #----------------------------------
    #Learn Method, takes a string (the new skill learned.
    def learn(self, newSkill):
        #Append the new skill to the skills attribute (the list)
        self.skills.append(newSkill)
    #Forget Method, doesn't take anything
    def forget(self):
        #Retrieve the last skill from the end of the list. It should be at position len(list)-1
        lastSkill = self.skills[(len(self.skills))-1]
        #If the Jedi doesn't have any skills, print "No Skills"
        if ((len(self.skills)) == 0):
            return 'Jedi has no skills!'
        #If the len of the skills list is bigger than 1
        else:
            #First print the last skill retrived
            print(lastSkill)
            #Then pop the last value from the skills list.
            self.skills.pop()
            #Return the last skill.
            return lastSkill
    #Check Skills Method, takes nothing
    def checkSkills(self):
        #The set function in python actually gets rid of repeating values and just gives you a list with unique values.
        findUnique = set(self.skills)
        #After we have a list of unique values, all we do is get the length of it.
        lenUnique = len(findUnique)
        #Return how many unique skills are present.
        return lenUnique
    #Duel Method, takes another jedi as an argument.
    def duel(self, other_jedi):
        #Take the difference in skills between the jedi and the other jedi. checkSkills already puts them into a unique list.
        differenceChecker = (self.checkSkills() - other_jedi.checkSkills())
        #If they don't have the same number of skills...
        if (differenceChecker != 0):
            #Find out which Jedi has more skills first.
            skillsMax = max(self.checkSkills(), other_jedi.checkSkills())
            #Using a self-if-else to check who the loser is. Loser is self if they don't have more skills. Otherwise loser is the other jedi.
            losingJedi = self if (self.checkSkills() != skillsMax) else other_jedi
            #Set losingJedi's boolean to true to make them a ghost.
            losingJedi.isGhost = True
            #Similarly using a self-if-else to check who the winner is. Winner is self if they have more skills. Otherwise winner is the other jedi.
            winningJedi = self if (self.checkSkills() == skillsMax) else other_jedi
            #Using format, print which Jedi won the duel.
            print("{} wins the duel.".format(winningJedi.name))
            #Return the winner's name.
            return winningJedi.name
        #If they have the same amount of unique skills...
        else:
            #Then it ends in a tie.
            print("The duel results in a tie.")
            return "Tie"
```

Implement a `Node` class with the following attributes and methods.

`Node(key, l, r, p)`
* `key`: an integer key to uniquely identify each node in a graph (input during instantiation)
* `left_child`: an integer key identifying the left child of the given node  (input `l` during instantiation)
* `right_child`: an integer key identifying the right child of the given node (input `r` during instantiation)
* `parent`: an integer key identifying the parent of the given node (input `p` during instantiation)
* `getChildren()`: returns a list of two elements, the left and right children the given node (in that order)

Note that you should appropriately specify default arguments to cover the cases where the node does not have a left/right child or a parent.

Also note that your code for `getChildren()` should return a list of two `Node` objects as the children, *not* their keys.

```python
class Node:
    #Takes a key, left (l), right (r), parent (p)
    def __init__(self, key, l=None, r=None, p=None):
        #Initialization of the variables
        self.key = key
        self.left_child = l
        self.right_child = r
        self.parent = p
    #The get children function, which returns left child and right child as a list.
    def getChildren(self):
        return list((self.left_child, self.right_child))
```

  * add the new node as the right child if the parent has a left child only
  * don’t add the node if the parent already has two children. Instead, print the message: "Parent has two children, node not added."
* if a new node is successfully added to the tree, then return `True`; otherwise, return `False`

```python
class Tree:
    
    def __init__(self, rootkey):
        #create a new tree while setting root
        self.root = Node(rootkey, None, None, None)

    def checkTree(self, key, parentKey, root):
        #Recursive function that searches through tree to find if parentKey exists
        # note that 'root' input is not necessarily the root of the tree ('self')
        # 'root' is just where to start looking for the right parentKey to add this new node
        if root == None:
            #if there is no root in tree
            return False
        if root.key == parentKey:
            if root.left_child == None or root.right_child == None:
                # the node 'root' is the parent you should add the new child node to
                return root 
            else:
                print("Parent has two children, node not added.")
                return False
        else:
            for child in root.getChildren():
                # check 'root' node's children if they are the parent you're looking for
                add_temp = self.checkTree(key, parentKey, child)
                if add_temp:
                    return add_temp

                
# YOUR CODE GOES HERE


    def add(self, key, parentKey):
        #First we need to make sure the parent exists.
        parent = self.checkTree(key, parentKey, self.root)
        #If the Parent exists, then we will need to check on how many children it has.
        if(parent != False):
            #If the parent has a left child, but doesn't have a right child, then add the right child.
            if(parent.left_child != None and parent.right_child == None):
                #Add the child to the right.
                parent.right_child = Node(key, None, None, parentKey)
                #Return true so we know it was successful.
                return True
            #If the parent has no children, add the child as a left node.
            elif(parent.left_child == None and parent.right_child == None):
                #Add the child to the left.
                parent.left_child = Node(key, None, None, parentKey)
                #Return true so we know it was successful.
                return True
            #If the parent already has both kids... then we can't add a node.
            else:
                print("Parent has two children, node not added.")
                return False
        #If the Parent doesn't exist, then we can't add a child, so we need to give error message.
        else:
            print("Parent not found.")
            return False

    def findNodeDelete(self, key, root):
        if root == None:
            return False
        if key == root.key:
            if root.left_child == None and root.right_child == None:
                if root.parent.left_child.key == key:
                    root.parent.left_child = None
                elif root.parent.right_child.key == key:
                    root.parent.right_child = None
                root = None
                return True
            else:
                print("Node not deleted, has children")
                return False
        else:
            for child in root.getChildren():
                delete_node = self.findNodeDelete(key, child)
                if delete_node:
                    return delete_node

    def delete(self, key):
        if self.root == None:
            self.root = Node(key, None, None, None)
        if key == self.root.key:
            if self.root.left_child == None and self.root.right_child == None:
                self.root = None
                return True
            else:
                print("Node not deleted, has children")
                return False
        else:
            for child in self.root.getChildren():
                delete_node = self.findNodeDelete(key, child)
                if delete_node:
                    return delete_node

        print("Parent not found." )
        return False

    def printTree(self):
        if self.root != None:
            print(self.root.key)
            for child in self.root.getChildren():
                self.printBranch(child)
        else: 
            return
        
    def printBranch(self, root):
        if root == None:
            return
        else:
            print(root.key)
            for child in root.getChildren():
                self.printBranch(child)
```
